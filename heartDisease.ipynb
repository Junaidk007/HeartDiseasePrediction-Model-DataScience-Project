### Preprocessing and Model traning

```python
df.dtypes
```

```python
# Encoding categorical features using Label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
# --- Encode Categorical Columns
label_encoders = {}

for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

print("\nAfter Encoding:")
df.head()
```

```python
# Split Features & Target
X = df.drop('target', axis=1)
Y = df['target']
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# Prediction and Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

Y_pred = model.predict(X_test)
print("Test Predictions:", Y_pred)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
print("Classification Report:")
print(classification_report(Y_test, Y_pred))
print("Confusion Matrix:")
print(confusion_matrix(Y_test, Y_pred))
```

```python
# Saving the model using pickle
import pickle

pickle.dump(model, open('model/heart_disease_model.pkl', 'wb'))
pickle.dump(scaler, open("model/scaler.pkl", "wb"))
pickle.dump(label_encoders, open("model/label_encoders.pkl", "wb"))
```

### Now testing on new data 

```python
model = pickle.load(open("model/heart_disease_model.pkl", "rb"))
scaler = pickle.load(open("model/scaler.pkl", "rb"))
label_encoders = pickle.load(open("model/label_encoders.pkl", "rb"))
```

```python
# New input
new_data_dict = {
    "age": 52,
    "sex": "Male",
    "chest_pain_type": "Typical angina",
    "resting_blood_pressure": 125,
    "cholestoral": 212,
    "fasting_blood_sugar": "Lower than 120 mg/ml",
    "rest_ecg": "Normal",
    "Max_heart_rate": 168,
    "exercise_induced_angina": "No",
    "oldpeak": 1.0,
    "slope": "Upsloping",
    "vessels_colored_by_flourosopy": 'Two',
    "thalassemia": "Reversable Defect"
}

new_df = pd.DataFrame([new_data_dict])
```

```python
# Encode
for col in new_df.columns:
    if col in label_encoders:
        new_df[col] = label_encoders[col].transform(new_df[col])

# Scale
new_scaled = scaler.transform(new_df)
```

```python
# Predict
prediction = model.predict(new_scaled)

print("Prediction:", prediction)

if prediction[0] == 1:
    print("⚠️ Heart Disease Detected")
else:
    print("✅ No Heart Disease")
```
